```python
# Map imgN.exr / imgN_gamma.exr -> original IDs and optionally convert to .hdr
import os, re, shutil, sys

# Paths (adjust if needed)
pix2pix_input_dir = "/content/Deep-HdrReconstruction/test_inputs_pix2pix"
exr_out_dir = "/content/Deep-HdrReconstruction/recon_results_from_pix2pix"
renamed_dir = os.path.join(exr_out_dir, "renamed_by_id")
os.makedirs(renamed_dir, exist_ok=True)

# If True, convert EXR -> HDR (Radiance). If False, keep .exr (just rename/move).
convert_to_hdr = True

# Helpers for reading/writing if conversion required
def read_exr_safe(path):
    # prefer pyexr (simple), fallback to OpenEXR
    try:
        import pyexr
        arr = pyexr.read(path).astype('float32')
        return arr
    except Exception:
        try:
            import OpenEXR, Imath
            exr = OpenEXR.InputFile(path)
            header = exr.header()
            dw = header['dataWindow']
            w = dw.max.x - dw.min.x + 1
            h = dw.max.y - dw.min.y + 1
            pt = Imath.PixelType(Imath.PixelType.FLOAT)
            chs = header.get('channels', {}).keys()
            if all(c in chs for c in ("R","G","B")):
                r = np.frombuffer(exr.channel('R', pt), dtype='float32').reshape(h, w)
                g = np.frombuffer(exr.channel('G', pt), dtype='float32').reshape(h, w)
                b = np.frombuffer(exr.channel('B', pt), dtype='float32').reshape(h, w)
                import numpy as np
                return np.stack([r,g,b], axis=-1)
            else:
                raise RuntimeError("OpenEXR: channels not RGB")
        except Exception as e:
            raise RuntimeError(f"Couldn't read EXR with pyexr/OpenEXR: {e}")

def write_hdr_safe(path, arr):
    # try imageio HDR writer, else fallback to PFM
    try:
        import imageio.v3 as iio
        iio.imwrite(path, arr.astype('float32'), plugin="HDR-FI")
        return True
    except Exception as e:
        # fallback: write PFM
        try:
            write_pfm(path.replace('.hdr','.pfm'), arr)
            return True
        except Exception as e2:
            print("Failed to write HDR/PFM:", e, e2)
            return False

def write_pfm(path, arr):
    import numpy as np
    arr = arr.astype('float32')
    h,w = arr.shape[:2]
    color = (arr.ndim == 3 and arr.shape[2] == 3)
    with open(path, 'wb') as f:
        header = b'PF\n' if color else b'Pf\n'
        f.write(header)
        f.write(f"{w} {h}\n".encode('ascii'))
        f.write(b"-1.0\n")  # little-endian
        # write data
        arr.tofile(f)

# Gather input IDs (preserve sorted order)
input_files = sorted([f for f in os.listdir(pix2pix_input_dir) if f.lower().endswith('.png')])
input_ids = [os.path.splitext(f)[0] for f in input_files]  # e.g., pix2pix_fake_9C4A...
print("Found", len(input_files), "pix2pix input files (in sorted order).")

# Gather EXR outputs
exr_files = sorted([f for f in os.listdir(exr_out_dir) if f.lower().endswith('.exr')])
print("Found", len(exr_files), "EXR output files:", exr_files[:10])

# Function to extract index from EXR filename
def extract_index(fname):
    # common patterns: img12.exr, IMG_12.exr, 12.exr, ... try several regexes
    patterns = [
        r'img[_\-]?0*([0-9]+)',     # img12, img_12, img-12
        r'IMG[_\-]?0*([0-9]+)',
        r'^0*([0-9]+)',             # starts with number
        r'[_\-]0*([0-9]+)[^0-9]*$', # trailing number before suffix
        r'([0-9]+)'                 # any number (last fallback)
    ]
    for p in patterns:
        m = re.search(p, fname, flags=re.IGNORECASE)
        if m:
            try:
                return int(m.group(1))
            except:
                pass
    return None

# Build mapping: try index-based mapping (img1 -> input_ids[0]), else positional mapping
mappings = []  # tuples of (src_exr, dst_basename, note)
used_inputs = set()

for i, exr in enumerate(exr_files):
    idx = extract_index(exr)  # 1-based if matched like img1 -> 1
    suffix = ""
    # detect gamma suffix
    if re.search(r'gamma', exr, flags=re.IGNORECASE):
        suffix = "_gamma"
    elif re.search(r'lin|linear', exr, flags=re.IGNORECASE):
        suffix = "_linear"
    # decide target input id
    target_id = None
    note = ""
    if idx is not None:
        # map img1 -> input index 1 -> input_ids[idx-1] if exists
        pos = idx - 1
        if 0 <= pos < len(input_ids):
            target_id = input_ids[pos]
            note = f"index {idx} -> positional input #{pos}"
        else:
            note = f"index {idx} out of range, will fallback to sequential mapping"
    if target_id is None:
        # fallback sequential mapping by available unused inputs in order
        # pick first unused input in order, else None
        for uid in input_ids:
            if uid not in used_inputs:
                target_id = uid
                note = "sequential mapping (first unused)"
                break
    if target_id is None:
        # final fallback use exr basename
        target_id = os.path.splitext(exr)[0]
        note = "no inputs available; using exr basename"

    used_inputs.add(target_id)
    # build destination basename
    dst_base = target_id + suffix
    mappings.append((exr, dst_base, note))

# Perform rename/conversion
for src_exr, dst_base, note in mappings:
    src_path = os.path.join(exr_out_dir, src_exr)
    if convert_to_hdr:
        # read then write .hdr
        try:
            arr = read_exr_safe(src_path)
            dst_path = os.path.join(renamed_dir, dst_base + ".hdr")
            ok = write_hdr_safe(dst_path, arr)
            if ok:
                print(f"Converted {src_exr} -> {os.path.basename(dst_path)}  ({note})")
            else:
                # fallback: copy original EXR but with new name
                fallback = os.path.join(renamed_dir, dst_base + ".exr")
                shutil.copy2(src_path, fallback)
                print(f"Conversion failed; copied EXR to {os.path.basename(fallback)} ({note})")
        except Exception as e:
            print(f"ERROR reading/converting {src_exr}: {e}. Will copy as EXR.")
            shutil.copy2(src_path, os.path.join(renamed_dir, dst_base + ".exr"))
    else:
        # just rename / copy EXR
        dst_path = os.path.join(renamed_dir, dst_base + ".exr")
        shutil.copy2(src_path, dst_path)
        print(f"Copied {src_exr} -> {os.path.basename(dst_path)}  ({note})")

print("\nDone. Renamed/converted files are in:", renamed_dir)
print("Sample files:", sorted(os.listdir(renamed_dir))[:20])
```

```python
# View EXR safely (uses pyexr or OpenEXR fallback)
import os, glob, numpy as np, matplotlib.pyplot as plt

OUT_DIR = "/content/Deep-HdrReconstruction/recon_results_from_pix2pix"
exr_files = sorted([os.path.join(OUT_DIR,f) for f in os.listdir(OUT_DIR) if f.lower().endswith('.exr')])
print("Found", len(exr_files), "EXR files")

def read_exr_pyexr(path):
    import pyexr
    arr = pyexr.read(path)   # returns (H,W,3) float32 or (H,W) if grayscale
    return arr.astype(np.float32)

def read_exr_openexr(path):
    import OpenEXR, Imath
    exr = OpenEXR.InputFile(path)
    header = exr.header()
    dw = header['dataWindow']
    w = dw.max.x - dw.min.x + 1
    h = dw.max.y - dw.min.y + 1
    # request as 32-bit floats
    pt = Imath.PixelType(Imath.PixelType.FLOAT)
    # Common channel names: R,G,B (or channels list from header)
    channels = header.get('channels', {}).keys()
    # prefer RGB if available
    wanted = ['R','G','B']
    present = [c for c in wanted if c in channels]
    if len(present) >= 3:
        r = np.frombuffer(exr.channel('R', pt), dtype=np.float32).reshape(h, w)
        g = np.frombuffer(exr.channel('G', pt), dtype=np.float32).reshape(h, w)
        b = np.frombuffer(exr.channel('B', pt), dtype=np.float32).reshape(h, w)
        arr = np.stack([r,g,b], axis=-1)
    else:
        # fallback: read all channels and stack in file-order
        ch_names = list(channels)
        chdata = []
        for c in ch_names:
            try:
                ch = np.frombuffer(exr.channel(c, pt), dtype=np.float32).reshape(h, w)
                chdata.append(ch)
            except Exception:
                pass
        if not chdata:
            raise RuntimeError("No readable channels in EXR.")
        arr = np.stack(chdata, axis=-1)
    return arr

def read_exr(path):
    # try pyexr first (simpler), then OpenEXR fallback
    try:
        return read_exr_pyexr(path)
    except Exception as e:
        # print(e)  # Uncomment to debug
        return read_exr_openexr(path)

def tonemap_for_display(hdr, pct_low=0.5, pct_high=99.5, gamma=1/2.2, eps=1e-9):
    # Collapse alpha if present
    if hdr.ndim == 3 and hdr.shape[2] > 3:
        hdr = hdr[..., :3]
    # If single channel -> replicate to 3
    if hdr.ndim == 2:
        hdr = np.stack([hdr,hdr,hdr], axis=-1)
    # robust min/max using percentiles
    mn = np.nanpercentile(hdr, pct_low)
    mx = np.nanpercentile(hdr, pct_high)
    if not np.isfinite(mn) or not np.isfinite(mx) or mx - mn < eps:
        # fallback to min/max
        mn = np.nanmin(hdr)
        mx = np.nanmax(hdr)
    img = np.clip((hdr - mn) / (mx - mn + eps), 0, 1)
    img = np.clip(img ** gamma, 0, 1)
    return img

# show first 3 images with previews
for path in exr_files[:3]:
    try:
        hdr = read_exr(path)
    except Exception as e:
        print("Failed to read", path, ":", e)
        continue

    print("File:", os.path.basename(path), " shape:", hdr.shape, " dtype:", hdr.dtype,
          " min/max:", np.nanmin(hdr), np.nanmax(hdr))

    vis = tonemap_for_display(hdr, pct_low=0.5, pct_high=99.5, gamma=1/2.2)

    plt.figure(figsize=(10,5))
    plt.imshow(vis)
    plt.title(os.path.basename(path))
    plt.axis('off')
    plt.show()

    # optionally save preview PNG next to EXR
    save_preview = True
    if save_preview:
        out_png = path.rsplit('.',1)[0] + "_preview.png"
        import imageio
        imageio.imwrite(out_png, (255 * vis).astype(np.uint8))
        print("Saved preview:", out_png)
```

```python
# Robust EXR -> Radiance .hdr converter (no FreeImage required)
import os, re, sys, traceback
import numpy as np

# Paths - adjust if needed
pix2pix_input_dir = "/content/Deep-HdrReconstruction/test_inputs_pix2pix"
exr_out_dir = "/content/Deep-HdrReconstruction/recon_results_from_pix2pix"
out_hdr_dir = os.path.join(exr_out_dir, "renamed_by_id_hdr")
os.makedirs(out_hdr_dir, exist_ok=True)

# Gather inputs and exrs
input_files = sorted([f for f in os.listdir(pix2pix_input_dir) if f.lower().endswith('.png')])
input_ids = [os.path.splitext(f)[0] for f in input_files]   # pix2pix_fake_...
exr_files = sorted([f for f in os.listdir(exr_out_dir) if f.lower().endswith('.exr')])

print("Found", len(exr_files), "EXR outputs and", len(input_files), "pix2pix inputs.")

# EXR reader (pyexr preferred, fallback OpenEXR)
def read_exr(path):
    try:
        import pyexr
        arr = pyexr.read(path).astype(np.float32)
        # ensure HxWxC with C=3
        if arr.ndim == 2:
            arr = np.stack([arr,arr,arr], axis=-1)
        if arr.shape[2] > 3:
            arr = arr[..., :3]
        return arr
    except Exception as e_py:
        # fallback to OpenEXR
        try:
            import OpenEXR, Imath
            import numpy as _np
            exr = OpenEXR.InputFile(path)
            header = exr.header()
            dw = header['dataWindow']
            w = dw.max.x - dw.min.x + 1
            h = dw.max.y - dw.min.y + 1
            pt = Imath.PixelType(Imath.PixelType.FLOAT)
            channels = list(header.get('channels', {}).keys())
            # prefer R,G,B
            if all(c in channels for c in ("R","G","B")):
                r = _np.frombuffer(exr.channel('R', pt), dtype=_np.float32).reshape(h, w)
                g = _np.frombuffer(exr.channel('G', pt), dtype=_np.float32).reshape(h, w)
                b = _np.frombuffer(exr.channel('B', pt), dtype=_np.float32).reshape(h, w)
                arr = _np.stack([r,g,b], axis=-1)
            else:
                # read up to three channels available
                chs = []
                for c in channels[:3]:
                    try:
                        chs.append(_np.frombuffer(exr.channel(c, pt), dtype=_np.float32).reshape(h, w))
                    except Exception:
                        pass
                if not chs:
                    raise RuntimeError("No readable channels in EXR")
                while len(chs) < 3:
                    chs.append(chs[-1])
                arr = _np.stack(chs[:3], axis=-1)
            return arr.astype(np.float32)
        except Exception as e_oe:
            raise RuntimeError(f"read_exr failed: pyexr_err={e_py}; OpenEXR_err={e_oe}")

# RGBE encoder (Paul Haeberli style)
def float_to_rgbe_image(img):
    # img: H x W x 3 float32 (RGB)
    H, W, C = img.shape
    assert C >= 3
    R = img[...,0]
    G = img[...,1]
    B = img[...,2]
    maxc = np.maximum(np.maximum(R, G), B)
    rgbe = np.zeros((H, W, 4), dtype=np.uint8)

    # mask for non-zero pixels
    nz = maxc > 1e-32
    if np.any(nz):
        # exponent e = floor(log2(maxc)) + 1
        e = np.zeros_like(maxc, dtype=np.int32)
        e[nz] = np.floor(np.log2(maxc[nz])).astype(np.int32) + 1
        # mantissa scaled to [0,255]
        scale = np.power(2.0, e - 8)  # because mantissa = val / 2^e * 256 => val / (2^(e-8))
        # compute mantissas
        rgbe[...,0][nz] = np.clip((R[nz] / scale[nz]).astype(np.int32), 0, 255)
        rgbe[...,1][nz] = np.clip((G[nz] / scale[nz]).astype(np.int32), 0, 255)
        rgbe[...,2][nz] = np.clip((B[nz] / scale[nz]).astype(np.int32), 0, 255)
        rgbe[...,3][nz] = np.clip((e[nz] + 128).astype(np.int32), 0, 255)
    # zeros remain for very dark pixels
    return rgbe

# Radiance header + raw pixels writer (uncompressed, simple)
def write_hdr_rgbe(path, img):
    # img: HxWx3 float32
    H, W = img.shape[:2]
    rgbe = float_to_rgbe_image(img)
    # Radiance header
    header = []
    header.append("#?RADIANCE")
    header.append("# Written by Colab RGBE writer")
    header.append("FORMAT=32-bit_rle_rgbe")  # many readers accept this even if no RLE data follows
    header.append("")  # blank line
    header.append(f"-Y {H} +X {W}\n")
    with open(path, "wb") as f:
        f.write(("\n".join(header)).encode('ascii'))
        # write raw scanline data in row-major order: R G B E per pixel
        # Many readers expect RLE; uncompressed raw following header may still be accepted by common viewers.
        # We'll write scanlines raw for maximum compatibility with simple readers.
        f.flush()
        # write bytes
        f.write(bytearray(rgbe.reshape(-1).tolist()))
    return True

# mapping function: extract index like img1 -> 1
def extract_index(fname):
    m = re.search(r'img[_\-]?0*([0-9]+)', fname, re.IGNORECASE)
    if m:
        return int(m.group(1))
    # fallback any trailing number before extension
    m2 = re.search(r'([0-9]+)(?=[^0-9]*\.[eE]xr$)|([0-9]+)(?=[^0-9]*$)', fname)
    if m2:
        for g in m2.groups():
            if g:
                return int(g)
    return None

# Build mapping for exr -> target base name
used = set()
mappings = []
for i, exr in enumerate(exr_files):
    idx = extract_index(exr)
    suffix = ""
    if re.search(r'gamma', exr, re.IGNORECASE):
        suffix = "_gamma"
    elif re.search(r'lin|linear', exr, re.IGNORECASE):
        suffix = "_linear"
    target = None
    note = ""
    if idx is not None:
        pos = idx - 1
        if 0 <= pos < len(input_ids):
            target = input_ids[pos]
            note = f"index {idx}"
    if target is None:
        # sequential pick first unused input
        for uid in input_ids:
            if uid not in used:
                target = uid
                note = "sequential"
                break
    if target is None:
        # fallback use exr basename
        target = os.path.splitext(exr)[0]
        note = "fallback basename"
    used.add(target)
    dst_base = target + suffix
    mappings.append((exr, dst_base, note))

# Convert and write
for exr_name, dst_base, note in mappings:
    exr_path = os.path.join(exr_out_dir, exr_name)
    out_name = dst_base + ".hdr"
    out_path = os.path.join(out_hdr_dir, out_name)
    print(f"Converting {exr_name} -> {out_name}  ({note})")
    try:
        img = read_exr(exr_path)  # float32 HxWx3
    except Exception as e:
        print("  ERROR reading EXR:", e)
        traceback.print_exc()
        continue
    try:
        write_hdr_rgbe(out_path, img)
        print("  WROTE HDR:", out_path)
    except Exception as e:
        print("  ERROR writing HDR:", e)
        traceback.print_exc()

print("\nDone. HDR files written to:", out_hdr_dir)
print("Sample files:", sorted(os.listdir(out_hdr_dir))[:40])
```
